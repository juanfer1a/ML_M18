# K Means - Generated data

Isotropic dataset module in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html) \
Train-test splits in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) \
k-Means module in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

```
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.datasets import make_gaussian_quantiles
from sklearn.model_selection import train_test_split
```

```
X, y = make_blobs(n_samples = 1000 , random_state = 170 )
# X, y = make_gaussian_quantiles(cov=2, n_samples=1000, n_classes=2, random_state=170)
# X[y==0] = X[y==0]*0.5

# Create train (80%) and test (20%) sets and print all shapes
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=170)

# Create Class plot and train/test plot
_, axes = plt.subplots(1,2, figsize=(8,4))
axes[0].scatter( X[:,0] , X[:,1], c=y, edgecolors='k', cmap="Paired")
axes[1].scatter( X_train[:,0], X_train[:,1], c='r', edgecolors='k')
axes[1].scatter( X_test[:,0], X_test[:,1], c='b', edgecolors='k')

print (X.shape, y.shape, X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```
kmeans = KMeans(n_clusters=3)

# Train Kmeans
kmeans.fit(X_train)
```

```
# Plot centroids
centroids = kmeans.cluster_centers_
print(centroids)
```

```
# Create predictions
predictions = kmeans.predict(X_test)
```

```
_, ax = plt.subplots (figsize = (5,4))
colors=['blue','cyan','magenta']

# plot data and found centroids
ax.scatter(X_test[:,0], X_test[:,1], c='b', edgecolors='k')
ax.scatter(centroids[:,0], centroids[:,1], facecolor="none", edgecolors='r', s=150, linewidth=3, marker="o")
```
