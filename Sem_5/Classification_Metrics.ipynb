# Regression

```python
# import useful libraries
import numpy as np
import matplotlib.pyplot as plt

# import data
from sklearn.datasets import make_regression

# import classifiers
from sklearn.linear_model import LinearRegression

# Import validation
from sklearn.model_selection import train_test_split
```

```python
# Create random data
X, y = make_regression(n_samples = 1000, n_features = 1 , noise = 20, random_state = 1)

# Split train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)

# added some noise for separate train/test samples
np.random.seed(0)
y_test = y_test + np.random.normal(len(y_test)) * 0.1

y_train, y_test = y_train/100, y_test/100

# Plot train / test data
_, axes = plt.subplots(1,2,figsize=(8,4))
axes[0].scatter(X, y, edgecolors='k')
axes[1].scatter(X_train, y_train, c='r', edgecolors='k')
axes[1].scatter(X_test, y_test, c='b', edgecolors='k')
```

```python
# Train a linear regression

linear_regression = LinearRegression().fit(X_train, y_train)

# print all results
print ("m: {0:.4f}, b: {1:.4f}".format(linear_regression.coef_[0], linear_regression.intercept_))
```

```python
# Create predictions
predictions = linear_regression.predict(X_test)
```

```python
_, ax = plt.subplots (figsize=(5,4))
ax.scatter(X_train, y_train, c='r', edgecolors='k')
ax.scatter(X_test, y_test, c='b', edgecolors='k')
ax.plot(X_test, predictions, c='g')
```

```python
linear_regression.coef_[0] = 0.3828
linear_regression.intercept_ = 0.0075

print ("m: {0:.4f}, b: {1:.4f}".format(linear_regression.coef_[0], linear_regression.intercept_))
```

```python
# Compute MAE and MSE
mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error (y_test, predictions)

# print everything
print ("MSE: {0:.4f}, MAE: {1:.4f}".format(mse, mae))
print (f"MSE: {mse:.4f}, MAE: {mae:.4f}")
```
