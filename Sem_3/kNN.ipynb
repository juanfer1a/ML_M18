# **K Nearest Neighbors**



Iris dataset [here](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris) \
k-NN documentation [here](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html#sklearn.neighbors.KNeighborsClassifier)

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier as kNN
from sklearn.datasets import load_iris # Iris dataset
```

```
iris = load_iris() # Load the iris dataset
iris_x = iris.data[:, 0:2] # Features. We'll pick 2
iris_y = iris.target # Labels
iris_names = iris.target_names
iris_features = iris.feature_names[0:2]

# Print shapes
print ( iris_x.shape, iris_y.shape, '\n', iris_names, '\n', iris_features)
```

```
_, ax = plt.subplots(figsize=(4,4))
ax.scatter(iris_x[:,0], iris_x[:,1], c=iris_y, cmap='Paired')
```

```
# Instance kNN object
n_neighbors = 3
knn = kNN(n_neighbors = n_neighbors , metric = "euclidean")

# Training
knn.fit( iris_x , iris_y ) # (num_samples, num_features), (num_samples)
```

```
## Plot decision boundaries
offset = 0.1
h_min, h_max = iris_x[:, 0].min()-offset, iris_x[:, 0].max()+offset
v_min, v_max = iris_x[:, 1].min()-offset, iris_x[:, 1].max()+offset

h_grid, v_grid = np.meshgrid(np.linspace(h_min, h_max, 300), np.linspace(v_min, v_max, 300))

predictions = knn.predict(np.c_[h_grid.ravel(), v_grid.ravel()])
predictions = predictions.reshape(h_grid.shape)
print (h_grid.shape, v_grid.shape, predictions.shape)
```

```
_, ax = plt.subplots (figsize=(6,4))

ax.pcolormesh(h_grid, v_grid, predictions, cmap='Paired')
# ax.contour(h_grid, v_grid, predictions, colors='k', linestyles='--')
ax.scatter(iris_x[:, 0], iris_x[:, 1], c=iris_y, edgecolors='k', cmap='Paired')

ax.set_xlabel(iris_features[0])
ax.set_ylabel(iris_features[1])
ax.set_xticks([]), ax.set_yticks([])
plt.tight_layout()
```
