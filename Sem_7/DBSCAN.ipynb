# **Density-based spatial clustering (DBSCAN)**

DBSCAN documentation [here](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html)

```python
import numpy as np
from matplotlib import pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.model_selection import train_test_split
```

```python
X, y = make_blobs(n_samples = 500, n_features = 2, random_state=3)

# Train test splits
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 25)

_, axes = plt.subplots(1,2, figsize = (8,4))
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap='Paired', edgecolors='k')
axes[1].scatter(X_train[:, 0], X_train[:, 1], c='b', edgecolors='k')
axes[1].scatter(X_test[:, 0], X_test[:, 1], c='r', edgecolors='k')
plt.show()
```

```python
dbscan = DBSCAN(eps= 1 , min_samples= 5 )

# Train DBSCAN
dbscan.fit(X_train)
```

```python
# Predictions
predictions = dbscan.labels_
print (predictions.shape)
```

```python
labels, counts = np.unique(predictions, return_counts=True)
print(labels, counts)
```

```python
_, axes = plt.subplots(1,2,figsize = (10,4))
axes[0].scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='Paired', edgecolors='k')
axes[1].scatter(X_train[:, 0], X_train[:, 1], c=predictions, cmap='Paired', edgecolors='k')
```

# **DBSCAN on real data**

Download the fruits data [here](https://drive.google.com/file/d/1M8tiAWDZclABJN1Meq9oEHCLJTNyHdpP/view?usp=share_link)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN
from sklearn.model_selection import train_test_split
```

```python
fruits_data = pd.read_csv("fruits_data_ext.csv")

X = fruits_data.iloc[:, [5,6]].values
y = fruits_data.iloc[:, [1]].values

X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.2, random_state=2)

plt.scatter( X[:,0], X[:,1], c=y, cmap='Paired', edgecolors='k')
print (X.shape, y.shape)
```

```python
dbscan = DBSCAN(eps= 0.5 , min_samples= 5 )
dbscan.fit( X_train )
```

```python
predictions = dbscan.labels_
print (predictions.shape)

labels, counts = np.unique(predictions, return_counts=True)
print (labels, counts)
```

```python
# Plot
_, axes = plt.subplots(1,2,figsize=(8,4))
axes[0].scatter(X_train[:, 0], X_train[:,1], c=y_train, edgecolors='k', cmap='Paired')
axes[1].scatter(X_train[:, 0], X_train[:,1], c=predictions, edgecolors='k', cmap='Paired')
```
